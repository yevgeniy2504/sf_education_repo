```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
```

```python
melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

```python
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area/sum_area
display(melb_df['AreaRatio'])
```

```python
import pandas as pd

countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})

# Расчёт плотности населения
countries_df['density'] = (countries_df['population'] * 1_000_000) / countries_df['area']

# Средняя плотность
average_density = round(countries_df['density'].mean(), 2)

print(average_density)
```

```python
import pandas as pd

# Предположим, melb_df уже загружен
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)  # преобразуем в datetime
melb_df['WeekdaySale'] = melb_df['Date'].dt.dayofweek  # создаем признак дня недели

# Считаем количество продаж в выходные (суббота=5, воскресенье=6)
weekend_count = melb_df[melb_df['WeekdaySale'].isin([5, 6])].shape[0]

print(weekend_count)
```

```python
def get_weekend(weekday):
    return 1 if weekday in (5, 6) else 0

melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)

# 1. Маска: выходные
weekend_mask = melb_df['Weekend'] == 1

# 2. Маска: цена не пропущена
price_notna_mask = melb_df['Price'].notna()

# 3. Применяем обе маски
weekend_sales = melb_df[weekend_mask & price_notna_mask]

# 4. Считаем среднюю цену
average_price_weekend = round(weekend_sales['Price'].mean())

# 5. Показываем результат
print(average_price_weekend)
```

```python
top_list_salers = melb_df['SellerG'].value_counts().nlargest(49).index


melb_df['SellerG_short'] = melb_df['SellerG'].apply(lambda x: x if x in top_list_salers else 'Other' )

nelson_series = melb_df['SellerG_short'] == 'Nelson'

min_price_Nelson = melb_df[nelson_series]['Price'].min()

other_series = melb_df['SellerG_short'] == 'Other'

min_price_other = melb_df[other_series]['Price'].min()

print(abs(min_price_Nelson / min_price_other))

```

```python
# создаём пустой список
unique_list = []
# пробегаемся по именам столбцов в таблице
for col in melb_df.columns:
    # создаём кортеж (имя столбца, число уникальных значений)
    item = (col, melb_df[col].nunique(), melb_df[col].dtypes) 
    # добавляем кортеж в список
    unique_list.append(item) 
# создаём вспомогательную таблицу и сортируем её
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique',  ignore_index=True)
# выводим её на экран
display(unique_counts)
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
max_unique_count = 150 # задаём максимальное число уникальных категорий
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: # проверяем условие
        melb_df[col] = melb_df[col].astype('category') # преобразуем тип столбца
display(melb_df.info())
```

```python
# Узнаем размер до преобразования
before_mem = melb_df.memory_usage(deep=True).sum() / (1024**2)  # в мегабайтах
print(f"До преобразования: {before_mem:.2f} MB")
```

```python
top_suburbs = melb_df['Suburb'].value_counts().nlargest(119).index
melb_df['Suburb'] = melb_df['Suburb'].apply(lambda x: x if x in top_suburbs else 'other')

melb_df['Suburb'] = melb_df['Suburb'].astype('category')

after_mem = melb_df.memory_usage(deep=True).sum() / (1024**2)  # в мегабайтах
print(f"После преобразования: {after_mem:.2f} MB")

memory_diff = before_mem - after_mem
print(f"Разница: {round(memory_diff, 1)} MB")
```
