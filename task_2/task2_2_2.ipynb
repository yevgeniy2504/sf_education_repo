```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data_fe.csv')
melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])  # преобразуем в datetime
melb_df['WeekdaySale'] = melb_df['Date'].dt.dayofweek  # создаем признак дня недели
melb_df['Quarter'] = melb_df['Date'].dt.quarter
melb_df['Quarter'].value_counts().sort_values()

```

```python
# Список исключений
exclude_cols = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

# Перебор всех столбцов
for col in melb_df.columns:
    if col not in exclude_cols and melb_df[col].nunique() < 150:
        melb_df[col] = melb_df[col].astype('category')
        
# Подсчёт столбцов с типом category
category_count = melb_df.select_dtypes(include='category').shape[1]

# Вывод результата
print(category_count)

melb_df.info()
```

```python
melb_df.sort_values(by='Price')
```

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
# Сортировка по убыванию столбца AreaRatio с одновременным сбросом индексов
df_sorted = melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True)

# Получение значения площади здания в строке 1558 и округление до целого числа
area_value = round(df_sorted.loc[1558, 'Area'])

# Вывод результата
print(area_value)
```
